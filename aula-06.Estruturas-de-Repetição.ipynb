## Exemplo Resolvido 1: Tabuada

**O problema:** Crie um programa que peça um número ao usuário e imprima a tabuada desse número de 1 a 10 usando o `for`.

```python
numero = int(input("Você quer a tabuada de qual número? "))

print(f"\n--- Tabuada do {numero} ---")
for i in range(1, 11):
    resultado = numero * i
    print(f"{numero} x {i:2d} = {resultado:2d}")
```

*Dica: O formato `{:2d}` dentro da f-string diz para o Python reservar 2 espaços na hora de imprimir aquele número. Isso ajuda a deixar os números alinhados bonitinhos na tela!*

## Exemplo Resolvido 2: Somatória com While

**O problema:** Peça números ao usuário repetidamente. O programa deve somar todos os números digitados e só deve parar (e mostrar o total) quando o usuário digitar o **número ZERO**.

```python
soma = 0
while True:
    valor = float(input("Digite um valor para somar (ou 0 para parar): "))

    if valor == 0:
        break # Sai do loop quando o usuário digitar zero

    soma = soma + valor

print(f"\nA soma total foi de: {soma}")
```

### Desafio 1: Somatória dos Pares
Peça um número `N` ao usuário. Use um loop `for` ou `while` para calcular e exibir a SOMA de todos os números PARES entre 1 e `N`.

```python
N = int(input("Digite o valor final N: "))

soma_pares = 0

for numero in range(1, N + 1):
    if numero % 2 == 0:
        soma_pares += numero

print(f"A soma dos números pares entre 1 e {N} é: {soma_pares}")

```

### Desafio 3.
Supondo que a população de um país A seja da ordem de 80000 habitantes com uma taxa anual de crescimento de 3% e que a população de B seja 200000 habitantes com uma taxa de crescimento de 1.5%. Faça um programa que calcule e escreva o número de anos necessários para que a população do país A ultrapasse ou iguale a população do país B, mantidas as taxas de crescimento.

```python
pop_a = 80000
pop_b = 200000
anos = 0

while pop_a < pop_b:
    pop_a = pop_a * 1.03
    pop_b = pop_b * 1.015
    anos += 1

print(f"Serão necessários {anos} anos.")

```

### Desafio 4.
Altere o programa anterior permitindo ao usuário informar as populações e as taxas de crescimento iniciais. Valide a entrada e permita repetir a operação.

```python
while True:

    pop_a = int(input("Digite a população do país A: "))
    pop_b = int(input("Digite a população do país B: "))

    taxa_a = float(input("Digite a taxa de crescimento de A (%): "))
    taxa_b = float(input("Digite a taxa de crescimento de B (%): "))

    if pop_a <= 0 or pop_b <= 0 or taxa_a <= 0 or taxa_b <= 0:
        print("Valores inválidos!")
        continue

    anos = 0

    while pop_a < pop_b:
        pop_a = pop_a * (1 + taxa_a / 100)
        pop_b = pop_b * (1 + taxa_b / 100)
        anos += 1

    print(f"Serão necessários {anos} anos.")

    resposta = input("Deseja repetir? (s/n): ")

    if resposta.lower() != "s":
        break
```

### Desafio 18.
Faça um programa que, dado um conjunto de N números, determine o menor valor, o maior valor e a soma dos valores.

```python
n = int(input("Quantos números deseja informar? "))

soma = 0

for i in range(n):
    numero = int(input("Digite um número: "))

    if i == 0:
        maior = numero
        menor = numero

    if numero > maior:
        maior = numero

    if numero < menor:
        menor = numero

    soma += numero

print(f"Maior valor: {maior}")
print(f"Menor valor: {menor}")
print(f"Soma dos valores: {soma}")
```

### Desafio 21.
Faça um programa que peça um número inteiro e determine se ele é ou não um número primo. Um número primo é aquele que é divisível somente por ele mesmo e por 1.

```python
numero = int(input("Digite um número inteiro: "))

primo = True

if numero < 2:
    primo = False
else:
    for i in range(2, numero):
        if numero % i == 0:
            primo = False
            break

if primo:
    print("É um número primo")
else:
    print("Não é um número primo")
```

### Desafio 22.
Altere o programa de cálculo dos números primos, informando, caso o número não seja primo, por quais número ele é divisível.

```python
numero = int(input("Digite um número inteiro: "))

divisores = []

for i in range(2, numero):
    if numero % i == 0:
        divisores.append(i)

if len(divisores) == 0 and numero > 1:
    print("É um número primo")
else:
    print("Não é um número primo")
    print("É divisível por:", divisores)
```

### Desafio 23.
Faça um programa que mostre todos os primos entre 1 e N sendo N um número inteiro fornecido pelo usuário. O programa deverá mostrar também o número de divisões que ele executou para encontrar os números primos. Serão avaliados o funcionamento, o estilo e o número de testes (divisões) executados.

```python
n = int(input("Digite um número: "))

divisoes = 0

for numero in range(2, n + 1):
    primo = True

    for i in range(2, numero):
        divisoes += 1

        if numero % i == 0:
            primo = False
            break

    if primo:
        print(numero)

print(f"Total de divisões realizadas: {divisoes}")
```

### Desafio 25.
Faça um programa que peça para n pessoas a sua idade, ao final o programa devera verificar se a média de idade da turma varia entre 0 e 25,26 e 60 e maior que 60; e então, dizer se a turma é jovem, adulta ou idosa, conforme a média calculada.

```python
n = int(input("Quantas pessoas há na turma? "))

soma_idades = 0

for i in range(n):
    idade = int(input("Digite a idade: "))
    soma_idades += idade

media = soma_idades / n

print(f"Média de idade: {media:.2f}")

if media <= 25:
    print("Turma jovem")
elif media <= 60:
    print("Turma adulta")
else:
    print("Turma idosa")

```

### Desafio 26.
Numa eleição existem três candidatos. Faça um programa que peça o número total de eleitores. Peça para cada eleitor votar e ao final mostrar o número de votos de cada candidato.

```python
eleitores = int(input("Digite o número total de eleitores: "))

candidato1 = 0
candidato2 = 0
candidato3 = 0

for i in range(eleitores):
    voto = int(input("Vote no candidato (1, 2 ou 3): "))

    if voto == 1:
        candidato1 += 1
    elif voto == 2:
        candidato2 += 1
    elif voto == 3:
        candidato3 += 1
    else:
        print("Voto inválido!")

print("\nResultado da eleição:")
print(f"Candidato 1: {candidato1} votos")
print(f"Candidato 2: {candidato2} votos")
print(f"Candidato 3: {candidato3} votos")

```

### Desafio 27.
Faça um programa que calcule o número médio de alunos por turma. Para isto, peça a quantidade de turmas e a quantidade de alunos para cada turma. As turmas não podem ter mais de 40 alunos.

```python
turmas = int(input("Digite a quantidade de turmas: "))

total_alunos = 0

for i in range(turmas):

    alunos = int(input(f"Quantidade de alunos da turma {i + 1}: "))

    while alunos > 40:
        alunos = int(input("Turma não pode ter mais de 40 alunos. Digite novamente: "))

    total_alunos += alunos

media = total_alunos / turmas

print(f"Média de alunos por turma: {media:.2f}")

```
